**Q:** Create a function that receives a dataframe and a column name and adds to the dataframe a new column (called **columnName_binary**) that has the value 1 if the record is above average in that column and 0 otherwise.

```python
def binary_creator(df, column_name):
    avg_col = df[column_name].mean()
    df[column_name+"_binary"] = (df[column_name] > avg_col).astype(int)
    return df
```

```python
binary_creator(heart_df, 'age').head(2)
```

**Q:** Create a function that receives a dataframe and a column name. The function should create a new column in the dataframe that contains the Z-scored values of that column.

                    Consider that *z = (x – μ) / σ*

```python
def z_scorer(df, column_name):
    avg_col = df[column_name].mean()
    std_col = df[column_name].std()
    df[column_name+'_z_scored'] = (df[column_name]-avg_col)/std_col
    
    return df
```

```python
z_scorer(heart_df, 'age').head(2)
```

**Q:** Create a function named **df_descriptor** that receives a dataframe and a name of a (binary) column.
This function should print the overview (average, std, median,etc...) of the remaining variables in both values
of the binary variable.

```python
def df_descriptor(df, binary_column_name):
    df0 = df[df[binary_column_name] == 0]
    df1 = df[df[binary_column_name] == 1]
    
    return df0.describe(), df1.describe()
```

```python
desc_df0, desc_df1 = df_descriptor(heart_df, 'smoking')
```
