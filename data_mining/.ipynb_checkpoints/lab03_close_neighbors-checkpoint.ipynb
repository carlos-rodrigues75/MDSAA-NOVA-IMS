# Distance Matrix

```python
# imports
from sklearn import datasets
import numpy as np
```

```python
# abstract
dataset = datasets.load_iris()
```

```python
# dictionary
dataset.keys()
```

```python
dataset["feature_names"]
```

```python
data = dataset["data"]
# data  # data is a numpy array data structure. Think of it as a matrix of data (or as an excel spreadsheet)
```

```python
data.shape
```

```python
# euclidean distance of 2 observations
obs1 = data[10]
obs2 = data[81]

sum((obs2 - obs1)**2)**0.5
```

```python
# initialize distance matrix. What will be its final shape?
dist = []

for s1 in data:
    dist_one_vs_all = []
    for s2 in data:
        distance = sum((s1 - s2)**2)**0.5
        dist_one_vs_all.append(distance)
    dist.append(dist_one_vs_all)
```

```python
# Build the distance matrix. Use 2 for loops, the append list method and the euclidean distance formula
```

```python
np.array(dist)[:5, :5]
```
