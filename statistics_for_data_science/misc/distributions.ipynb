The PMF of the binomial distribution receive the following parameters:
- The number of "chances"
- The probability of each change
- The value you need to get the probability

The code below is used to answer the following question: 
- Suppose there are twelve multiple choice questions in an English
class quiz.
- Each question has five possible answers, and only one of them is
correct.
- Find the probability of having four or less correct answers if a
student attempts to answer every question at random.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n = 12
p = 0.2

r_values = list(range(n +1))
dist = [binom.pmf(r, n, p) for r in r_values]

bars = plt.bar(r_values, dist)
bars[4].set_color('g')

print("% of getting 4 questions right = ", binom.pmf(4, n, p))
plt.show()
```

# Normal Distribution

## Density Function:

$\frac{1}{\sigma \sqrt{ 2\pi}}e^{-\frac{1}{2}\left( \frac{x-\mu}{\sigma} \right)}$

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # Set the random seed for reproducibility
data = np.random.normal(5, 3, 10000)

mean = np.mean(data)
std = np.std(data)

plt.hist(data, bins=30, density=True, alpha=0.7)

# Plot the mean and standard deviations
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(mean - std, color='g', linestyle='dashed', linewidth=1, label='1 STD')
plt.axvline(mean + std, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mean - 2*std, color='b', linestyle='dashed', linewidth=1, label='2 STD')
plt.axvline(mean + 2*std, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mean - 3*std, color='m', linestyle='dashed', linewidth=1, label='3 STD')
plt.axvline(mean + 3*std, color='m', linestyle='dashed', linewidth=1)

plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('68-95-99.7 Rule')
plt.show()
```

We can use the following formula to convert the Normal Distribution to a Standard Normal Distribution (or Z-Distribution):
    

$$Z=\frac{x-\mu}{\sigma}$$

```python
data_standard = (data - mean) / std

plt.hist(data_standard, bins=30, density=True, alpha=0.7)

standard_mean = data_standard.mean()
standard_std = data_standard.std()

# Plot the mean and standard deviations
plt.axvline(standard_mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(standard_mean - standard_std, color='g', linestyle='dashed', linewidth=1, label='1 STD')
plt.axvline(standard_mean + standard_std, color='g', linestyle='dashed', linewidth=1)
plt.axvline(standard_mean - 2*standard_std, color='b', linestyle='dashed', linewidth=1, label='2 STD')
plt.axvline(standard_mean + 2*standard_std, color='b', linestyle='dashed', linewidth=1)
plt.axvline(standard_mean - 3*standard_std, color='m', linestyle='dashed', linewidth=1, label='3 STD')
plt.axvline(standard_mean + 3*standard_std, color='m', linestyle='dashed', linewidth=1)

plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('68-95-99.7 Rule')
plt.show()
```

# Z-Scores

Z-Scores are used to calculate the Standard Normal Distribution, but can be used to random variables to describe how far from the main the given random variable is in terms of standard deviation.

For instance, using the Normal Distribution above ($\mu = 5, \sigma = 3$):

$Z_5 = \frac{5-5}{3} = 0$

$Z_8 = \frac{8-5}{3} = 1$

$Z_2 = \frac{2-5}{3} = -1$

$Z_{6.5} = \frac{6.5-5}{3} = 0.5$ 


Z-Scores also give us a sense of the area under the probability distribution, for instance, we know that the likelihood to find a value below the mean is 50%, or the likelihood to find a value less than or equal to 1 standard deviation is ~84%. On the other hand, we can assume that the likelihood to find a value greater than 1 standard deviation is ~1-84%, which is ~16%.
